```python
import os
import imageio 
import glob
import json
```

```python
cnt=0

def point_tup(point_list):
    return (point_list[0][0],point_list[0][1],point_list[1][0],point_list[1][1])


def bbox_list(box_list):
    '''
        takes data['shapes'] as input 
    '''
    ret_list=[]
    for i in box_list:
        ret_list.append(point_tup(i['points']))
    return ret_list

for i in glob.glob("./*.json"):
    fname = "."+i.split('.')[1]+".txt"
    file = open(fname,"w")
    print(fname)
    with open(i) as json_file:  
        data = json.load(json_file)
        lst = bbox_list(data['shapes'])
        print(lst)
        for j in lst:
            temp=""
            temp+= "0 "
            temp+= str(min(j[0],j[1])/416)
            temp+= " "
            temp+= str(min(j[2],j[3])/416)
            temp+= " "
            temp+= str(abs(j[1]-j[0])/416)
            temp+= " "
            temp+= str(abs(j[3]-j[2])/416)
            temp+="\n"
            file.write(temp)
    file.close()
            
```
